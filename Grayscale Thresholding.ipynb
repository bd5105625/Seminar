```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
from matplotlib import cm
```

# Image treshold

```python
abc = []
path = 'image/ILD/1'#folder path
allFileList = os.listdir(path)#file under the folder
```

```python
imglist = []
filename = r'jpg'
# print(allFileList)
for file in allFileList:
    if file.endswith(filename):
        imglist.append(file)
#         print(file)
print(len(imglist))
```

```python
for file in imglist:
    img = cv2.imread(path + "/" + file)
    ret,thresh1 = cv2.threshold(img,200,255,cv2.THRESH_BINARY)
    temppath = path
    path = path + '/'+ 'gray'
    if not os.path.isdir(path):
        os.mkdir(path)
    distance = path + '/' + file.split('.')[0] + '_gray.jpg'
    path = temppath
    print(distance)
    cv2.imwrite(distance , thresh1)
#     cv2.imwrite((path + 'gray/' + file.split('.')[0] + '_gray.jpg'), )
# print(path + 'gray/' + file.split('.')[0] + '_gray.jpg')
# print(distance)
# cv2.imwrite("image/ILD/06753077/gray/ii.jpg" , thresh1)
plt.imshow(thresh1)
```

# Image Thresholding in different type

```python
img = cv2.imread('image/ILD/1/FILE1.jpg', 0)
# img = cv2.imread('image/CT-7358-0009.' , 0)
ret,thresh1 = cv2.threshold(img,200,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(0 , 6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
# plt.imshow(thresh1 , 'gray')
```

```python
img = cv2.medianBlur(img , 5)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,-3)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
titles = ['Original Image' , 'Normal Thresholding' , 'Mean Thresholding' , 'Gaussian Thresholding']
image = [img , thresh1 , th2 , th3]

for i in range(0 , 4):
    plt.subplot(2,2,i+1),plt.imshow(image[i] , 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    
plt.show()
```
